```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df.head(5)
```

```python
def clean_job_skills(val):
    if pd.notna(val):
        return ast.literal_eval(val)

df['job_skills'] = df['job_skills'].apply(clean_job_skills)
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded
```

```python
df_filtered = df_exploded.groupby(by=['job_title_short', 'job_skills'])['job_skills'].value_counts().to_frame()
```

```python
df_filtered
```

```python
df_filtered.reset_index(inplace= True)
```

```python
top = 3
```

```python
df_filtered.sort_values(by= 'count', ascending= False, inplace= True)
top_list = list(df_filtered.job_title_short.unique())
top_title = top_list[:top]
top_title
```

```python
df_filtered
```

```python
df_filtered[df_filtered['job_title_short'] == 'Data Scientist']
```

```python
fig, axs = plt.subplots(len(top_title), 1)
for i, titl in enumerate(top_title):
        df_plot = df_filtered[df_filtered['job_title_short'] == titl].head().copy()
        df_plot.plot(
                kind= 'barh',
                x= 'job_skills', 
                y= 'count',
                title= titl,
                xlabel= '', 
                ylabel= '',
                ax= axs[i],
                legend= False,
                xlim= (0, 120000),
        )
        axs[i].invert_yaxis()
fig.suptitle("In-Demand Job Skills", x = 0.55)
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(len(top_title), 1)

for i, titl in enumerate(top_title):
        df_plot = df_filtered[df_filtered['job_title_short'] == titl].head().copy()
        sns.barplot(
                data= df_plot,
                x= 'count', 
                y= 'job_skills',
                ax= axs[i],
                legend= False,
                hue= 'job_skills',
                palette= 'light:b_r'
        )
        axs[i].set_title(titl)
        axs[i].set_xlabel('')
        axs[i].set_ylabel('')
        axs[i].set_xlim(0, 120000)
        sns.despine(ax=axs[i])
fig.suptitle("In-Demand Job Skills", x = 0.55)
sns.set_theme(style='ticks')
plt.tight_layout()
plt.show()
```

```python
mapping = df['job_title_short'].value_counts().to_frame().copy()
mapping.reset_index(inplace= True)
mapping
```

```python
df_filtered= df_filtered.merge(mapping, how='left', left_on='job_title_short', right_on='job_title_short')
```

```python
df_filtered['Pecent'] = (df_filtered['count_x'] / df_filtered['count_y']) * 100
```

```python
df_filtered
```

```python
fig, axs = plt.subplots(len(top_title), 1)
for i, titl in enumerate(top_title):
        df_plot = df_filtered[df_filtered['job_title_short'] == titl].head().copy()
        sns.barplot(
                data= df_plot,
                x= 'Pecent', 
                y= 'job_skills',
                ax= axs[i],
                legend= False,
                hue= 'Pecent',
                palette= 'light:b'
        )
        for n, v in enumerate(df_plot['Pecent']):
                axs[i].text(x= v+1,y= n, s=f'{v:.0f}%', va='center')
        axs[i].set_title(titl)
        axs[i].set_xlabel('')
        axs[i].set_ylabel('')
        axs[i].set_xlim(0, 80)
        sns.despine(ax=axs[i])
sns.set_theme(style='ticks')
fig.suptitle("In-Demand Job Skills", x = 0.55)
plt.tight_layout()
plt.show()
```
