```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df.head(5)
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def clean_job_skills(val):
    if pd.notna(val):
        return ast.literal_eval(val)

df['job_skills'] = df['job_skills'].apply(clean_job_skills)
```

```python
job = 'Data Analyst'
country = 'United States'
```

```python
df_filtered = df[(df['job_title_short'] == job) & (df
                 ['job_country'] == country)].copy()
df_filteredjob_title_short
```

```python
df_exploded = df_filtered.explode('job_skills').copy()
```

```python
df_exploded
```

```python
df_exploded.loc[:, 'job_posted_month'] = df_exploded['job_posted_date'].dt.month
df_exploded
```

```python
df_pivot = df_exploded.pivot_table(index= 'job_posted_month', columns='job_skills',values= 'company_name', aggfunc='count', fill_value= 0)
```

```python
df_pivot
```

```python
df_pivot.loc['Total'] = df_pivot.sum()
```

```python
df_pivot
```

```python
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending = False).index].copy()
```

```python
df_pivot
```

```python
df_pivot.drop('Total', inplace= True)
```

```python
df_pivot
```

```python
df_pivot.iloc[:, :5].plot(
    kind= 'line',
    legend = True,
    title = 'Skills Posted in Job Requirements',
    ylabel = 'Skill Count',
    xlabel = ''
)
```

```python
df_filtered['job_posted_month'] = df_filtered['job_posted_date'].dt.month
```

```python
df_total = df_filtered.groupby(by= 'job_posted_month').size()
```

```python
df_plot = df_pivot.div(df_total / 100, axis= 0)
```

```python
df_plot
```

```python
df_plot.reset_index(inplace= True)
df_plot['job_posted_month'] = df_plot['job_posted_month'].apply(lambda x : pd.to_datetime(x, format= "%m").strftime("%b"))
df_plot.set_index('job_posted_month', inplace= True)
```

```python
from adjustText import adjust_text
from matplotlib.ticker import PercentFormatter
```

```python
plt.figure(figsize= (10, 6))
sns_plot= df_plot.iloc[:, :5]
sns.lineplot(
    data = sns_plot,
    dashes= False,
    legend= True,
    palette= 'hls',
)
texts = []
for i in range(5):
    texts.append(plt.text(x= 11.4, y= df_plot.iloc[-1, i], s=df_plot.columns[i]))
adjust_text(texts, arrowprops=dict(arrowstyle="->", color='black', lw=0.5))
plt.ylabel("Percentage")
plt.xlabel("")

axs = plt.gca()
axs.yaxis.set_major_formatter(PercentFormatter(decimals= 0))
sns.despine()
plt.tight_layout()
plt.title("Skills Posted in Job Requirements")
plt.show()
```
